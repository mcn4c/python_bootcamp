# Lists

```python
# declaring a list of numbers
nums = [5, 10, 15.2, 20]
print(nums)
#accessing elements within a list
print(nums[1])
print(nums[0])
print(nums[2] + nums[3])

#declaring a list of mixed data types
num = 4.3
data = [num, "word", True]
print(data)
#understanding lists within lists
data = [5, "book", [34, "hello"], True]
print(data)
print(data[2])
#using double bracket notation to access lists within lists
print(data[2][0])
inner_list= data[2]
print(inner_list[1])
print(data[1][3])# book is a string so it can also be indexed again, int and booleans cannot

#Changing values in a list using index
data = [5, 10, 15, 20]
print(data)
data[0] = 100
print(data)
a = [5, 10]
print(id(a))
a = [5, 10, 15]
print(id(a))# has different id from previous a
b = a
a = [5, 10]
b = a
print(id(b)) # b has same id as a

#understanding how lists are stored
a = [5, 10]
b = a
print(f"a: {a} \t b: {b}")
print(f"Location a[0]: {id(a[0])}\t Location b[0]: {id(b[0])}") #location of item in list is same
a[0] = 20 #redeclaring a[0] also changes b[0]
print(f"a: {a} \t b: {b}")

# using [:] to copy a list
data = [5, 10, 15, 20]
data_copy = data[:] #single colon copies the list can also use .copy() method

print(data)
print(data_copy) # same as data because it was copied
print(id(data))
print(id(data_copy))#different location

data[0] = 20
print(data)
print(data_copy) # did not change first index to 20 because it is just a copy, different location

sports = ["baseball", "basketball", "soccer", "swimming", "volleyball"]
print(f"I like to play {sports[0]}")
print(f"I like to play {sports[1]}")
print(f"I like to play {sports[2]}")
print(f"I like to go {sports[3]}")
print(f"I like to play {sports[4]}")

names = ["John", "Abraham", "Sam", "Kelly"]
print(f"{names[0][0]}\n{names[1][0]}\n{names[2][0]}\n{names[3][0]}")

```

# For Loops

```python
#writing first for loop using range
for num in range(5):
    print(f"Value: {num}")
# providing start, stop and step of range
for num in range(3, 13, 3):
    print(num)
#printing all characters in the name using the 'in' keyword
name = "John Smith"
for letter in name:
    print(f"Value: {letter}")

#using continue statement within a for loop
for num in range(5):
    if num == 3: # loop will skip over num = 3 because it will go to next iteration of loop before hitting print(num)
        continue
    print(num)
    
for num in range(5):
    if num == 3:
        break
    print(num)

#setting a placehoder using the 'pass' keyword
for i in range(5):
    #TODO: add code to print number
    pass

for i in range(0, 100, 3):
    print(i)

word = input("Give me any word: ").lower()

for letter in word:
    if letter == "a" or letter == "e" or letter == "i" or letter == "o" or letter == "u":
        print(letter)
    else:
        continue
```

# While Loops

```python
#writing first while loop
health = 10
while health > 0:
    print(health)
    health -= 1 # forgetting this line will result in an infinite loop

## general rule of thumb: counting with for loops, conditions with while loops

#using two or more loops together is called a nested loop
for i in range(3):
    for j in range(3):
        print(i,j)
        
#user input quit

word = input("Type a word infinitely, until you type quit ").lower()

while word != "quit":
        word = input("Type in a word: ").lower()
        
#Double loop practice

game_over = False
while not game_over:
    for num in range(5):       
        if num == 3:
            game_over = True
            break
        print(num)
```

# Working With Lists

```python
#checking the number of items within a list len() function
nums = [5, 10, 15]
length = len(nums)
print(length)

#accessing specific items of a list with slices
print(nums[1 : 3])
print(nums[:2]) # will put out items in index 0 and 1
print(nums[::2]) #will print every other index - 0, 2, 4, etc
print(nums[-2:]) #will print last two items in a list

# adding an item to the back of a list using append
nums = [10, 20]
nums.append(5)
print(nums)

#adding a value to the beginning of the list
words = ["ball", "base"]
print(words)
words.insert(0, "glove")
print(words)

#using pop to remove items and saving a variable to use later
items = [5, "ball", True]
items.pop()
print(items)
removed_item = items.pop(1)
print(removed_item, "\n", items)

#using the remove method with a try and except
sports = ["baseball", "soccer", "football", "hockey"]
print(sports)
try:
    sports.remove("soccer")
except:
    print("That item does not exist in the list")

print(sports)

#using min, max, and sum
nums = [5, 3, 9]
print(min(nums))
print(max(nums))
print(sum(nums))

#using sorted on lists for numerical and alphabetical data
nums = [5, 8, 0, 2]
sorted_nums = sorted(nums)
print(nums, sorted_nums)

#sorting a list with .sort in-place
nums = [5, 8, 0, 3]
nums.sort() #alters the original variable directly
print(nums)

#use conditional statements on a list
names = ["Jack", "Mary", "Robert"]
if "Mary" in names:
    print("found")
if "Jimmy" not in names:
    print("not found")

#using conditionals to check if a list is empty
nums = []
if not nums:  #could also say nums == []
    print("empty")
if nums:
    print("not empty")

#using a for loop to print all items in a list

sports = ["Baseball", "Hockey", "Football", "Basketball"]
for item in sports: #can call "item" anything, can use "sport"
    print(item)
    
#using the while loop to remove a certain value
names = ["Bob", "Jack", "Rob", "Bob", "Robert"]
print(names)
names.remove("Bob") # only removes one instance of Bob
print(names)

names = ["Bob", "Jack", "Rob", "Bob", "Robert"]
while "Bob" in names:
    names.remove("Bob") #removes all instances of Bob
print(names)

#remove duplicates using .count method

######### Code Below is not working, and I'm not sure why

names = ["Bob", "Kenny", "Amanda", "Bob", "Kenny"]

names2 = []
for i in names:
    if i not in names2:
        names2.append(i)
print(names2)
    




# User input into an array


user_input = ""
user_list = []
user_input = input("Type in a word here: ").lower()
while user_input != "quit":
   user_input = input("Type in a word here: ").lower()
    user_list.append(user_input)
print(user_list)
    
    
```

# Friday Project: Creating Hangman

```python
#import additional functions
from random import choice
from IPython.display import clear_output

#declare game variables
words = ["tree", "basket", "chair", "paper", "python"]
word = choice(words) #randomly chooses word from words list
guessed, lives, game_over = [], 7, False 

#create a list of underscores to match the length of the word chosen
guesses = ["_"] * len(word)

#Create main game loop
while not game_over:
    #output game information
    hidden_word = " ".join(guesses)
    print(f"Word to guess: {hidden_word}")
    print(f"Lives: {lives}")
    ans = input("Type quit or guess a letter: ").lower()
    if ans == "quit":
        print("Thanks for playing")
        game_over = True
```
