# Lists

```python
# declaring a list of numbers
nums = [5, 10, 15.2, 20]
print(nums)
#accessing elements within a list
print(nums[1])
print(nums[0])
print(nums[2] + nums[3])

#declaring a list of mixed data types
num = 4.3
data = [num, "word", True]
print(data)
#understanding lists within lists
data = [5, "book", [34, "hello"], True]
print(data)
print(data[2])
#using double bracket notation to access lists within lists
print(data[2][0])
inner_list= data[2]
print(inner_list[1])
print(data[1][3])# book is a string so it can also be indexed again, int and booleans cannot

#Changing values in a list using index
data = [5, 10, 15, 20]
print(data)
data[0] = 100
print(data)
a = [5, 10]
print(id(a))
a = [5, 10, 15]
print(id(a))# has different id from previous a
b = a
a = [5, 10]
b = a
print(id(b)) # b has same id as a

#understanding how lists are stored
a = [5, 10]
b = a
print(f"a: {a} \t b: {b}")
print(f"Location a[0]: {id(a[0])}\t Location b[0]: {id(b[0])}") #location of item in list is same
a[0] = 20 #redeclaring a[0] also changes b[0]
print(f"a: {a} \t b: {b}")

# using [:] to copy a list
data = [5, 10, 15, 20]
data_copy = data[:] #single colon copies the list can also use .copy() method

print(data)
print(data_copy) # same as data because it was copied
print(id(data))
print(id(data_copy))#different location

data[0] = 20
print(data)
print(data_copy) # did not change first index to 20 because it is just a copy, different location

sports = ["baseball", "basketball", "soccer", "swimming", "volleyball"]
print(f"I like to play {sports[0]}")
print(f"I like to play {sports[1]}")
print(f"I like to play {sports[2]}")
print(f"I like to go {sports[3]}")
print(f"I like to play {sports[4]}")

names = ["John", "Abraham", "Sam", "Kelly"]
print(f"{names[0][0]}\n{names[1][0]}\n{names[2][0]}\n{names[3][0]}")

```

# For Loops

```python
#writing first for loop using range
for num in range(5):
    print(f"Value: {num}")
# providing start, stop and step of range
for num in range(3, 13, 3):
    print(num)
#printing all characters in the name using the 'in' keyword
name = "John Smith"
for letter in name:
    print(f"Value: {letter}")

#using continue statement within a for loop
for num in range(5):
    if num == 3: # loop will skip over num = 3 because it will go to next iteration of loop before hitting print(num)
        continue
    print(num)
    
for num in range(5):
    if num == 3:
        break
    print(num)

#setting a placehoder using the 'pass' keyword
for i in range(5):
    #TODO: add code to print number
    pass

for i in range(0, 100, 3):
    print(i)

word = input("Give me any word: ").lower()

for letter in word:
    if letter == "a" or letter == "e" or letter == "i" or letter == "o" or letter == "u":
        print(letter)
    else:
        continue
```
